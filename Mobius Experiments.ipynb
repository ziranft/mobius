```python
import numpy as np
import matplotlib.pyplot as plt

u = np.linspace(-np.pi, np.pi, endpoint=True, num=25) 
v = np.linspace(-np.pi, np.pi, endpoint=True, num=25) 
x = 1 + np.cos(u)*np.cosh(v) 
y = np.sin(u)*np.sinh(v) 
z = np.sin(u)*np.cosh(v) 
ax = plt.axes(projection='3d')
ax = ax.plot_trisurf(x, y, z, linewidth=0, antialiased=False)
plt.show()
```
