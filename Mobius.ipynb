```
import numpy as np
import matplotlib.pyplot as plt

# define a Mobius strip
def mobius(u, v):
    x = (1 + v/2 * np.cos(u/2)) * np.cos(u)
    y = (1 + v/2 * np.cos(u/2)) * np.sin(u)
    z = v/2 * np.sin(u/2)
    return np.array([x, y, z])

# plot a Mobius strip
u = np.linspace(0, 2*np.pi, endpoint=True, num=100)
v = np.linspace(-1, 1, endpoint=True, num=11)
u, v = np.meshgrid(u, v)

mobius_strip = np.zeros((u.shape[0], u.shape[1], 3))
for i in range(u.shape[0]):
    for j in range(u.shape[1]):
        mobius_strip[i, j, :] = mobius(u[i, j], v[i, j])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(mobius_strip[:, :, 0], 
                mobius_strip[:, :, 1], 
                mobius_strip[:, :, 2], 
                rstride=1, 
                cstride=1, 
                color='b', 
                edgecolor='k')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# define a Mobius strip
def mobius(u,v):
    x = (1 + v/2*np.cos(u/2))*np.cos(u)
    y = (1 + v/2*np.cos(u/2))*np.sin(u)
    z = v/2*np.sin(u/2)
    return np.array([x,y,z])

# create parameter space
u = np.linspace(0, 2*np.pi, endpoint=True, num=100)
v = np.linspace(-0.5, 0.5, endpoint=True, num=10)

# create point grid
U, V = np.meshgrid(u, v)

# calculate mobius strip coordinates
X, Y, Z = mobius(U, V)

# plot
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
plt.show()
```
